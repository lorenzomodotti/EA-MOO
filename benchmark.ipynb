```python
from numpy import sqrt, sum, linspace, zeros

def f1(x):
    return x[0]

def f2_1(x):
    g = 1 + (9 / (len(x)-1)) * (sum(x) - x[0])
    return g * (1 - sqrt(x[0] / g))

def f2_2(x):
    g = 1 + (9 / (len(x)-1)) * (sum(x) - x[0])
    return [ x[0], g * (1 - (x[0] / g)**2) ]

###
# ZDT1
###
ZDT1_domain = [[0,1]] * 30

ZDT1 = [f1, f2_1]

def ZDT1_front(n):
    grid_1 = linspace(0, 1, n)
    grid_j = zeros(29)
    x = []
    for element in grid_1:
        x_k = []
        x_k.append(element)
        x_k.extend(grid_j)
        x.append(x_k)
    F1 = []
    F2 = []
    for x_k in x:
        F1.append(f1(x_k))
        F2.append(f2_1(x_k))
    return F1, F2

###
# ZDT2
###
ZDT2_domain = [[0,1]] * 30

ZDT2 = [f1, f2_2]

def ZDT2_front(n):
    grid_1 = linspace(0, 1, n)
    grid_j = zeros(29)
    x = []
    for element in grid_1:
        x_k = []
        x_k.append(element)
        x_k.extend(grid_j)
        x.append(x_k)
    F1 = []
    F2 = []
    for x_k in x:
        F1.append(f1(x_k))
        F2.append(f2_2(x_k))
    return F1, F2
```
