```python
import matplotlib.pyplot as plt
from EA.mooproblem import MOOProblem
from DE.main_de import de
from NSGA2.main_nsga2 import nsga2
from CMAES.main_cmaes import cmaes
from benchmark import *

N = 100
T = 250
problem = MOOProblem(ZDT1, ZDT1_domain)
F1, F2 = ZDT1_front(201)
```

```python
###
# DE
###

F = 0.5
CR = 0.1

de_solution = de(problem, N, T, F, CR)
de_front = de_solution.pareto_front()

plt.plot(F1, F2, c="black", label="Optimal Pareto Front")
plt.scatter(de_front[0], de_front[1], c="orange", marker=".", label="DE")
plt.legend(loc="upper right")
plt.title("ZDT1 | N=%d" %N + " | T=%d" %T)
plt.show()
```

```python
###
#NSGA 2
###

crossover_probability = 0.9
crossover_index = 20 
mutation_probability = 1/N 
mutation_index = 20

nsga2_solution = nsga2(problem, N, T, crossover_probability, crossover_index, mutation_probability, mutation_index)
nsga2_front = nsga2_solution.pareto_front()

plt.plot(F1, F2, c="black", label="Optimal Pareto Front")
plt.scatter(nsga2_front[0], nsga2_front[1], c="orange", marker=".", label="NSGA-II")
plt.legend(loc="upper right")
plt.title("ZDT1 | N=%d" %N + " | T=%d" %T)
plt.show()
```

```python
###
# CMA-ES
###

l=1
treshold=0.44

cmaes_solution = cmaes(problem, N, T, l, treshold)
cmaes_front = cmaes_solution.pareto_front()

plt.plot(F1, F2, c="black", label="Optimal Pareto Front")
plt.scatter(cmaes_front[0], cmaes_front[1], c="orange", marker=".", label="CMA-ES")
plt.legend(loc="upper right")
plt.title("ZDT1 | N=%d" %N + " | T=%d" %T)
plt.show()
```
